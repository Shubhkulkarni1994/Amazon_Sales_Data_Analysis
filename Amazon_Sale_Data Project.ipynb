# EDA

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,8),dpi=200)
ax=sns.countplot(y ='Item Type',data=df)
```

```python
plt.figure(figsize=(15,15),dpi=500)
sns.barplot(x='Total Revenue',y='Region',data=df)
```

```python
revenue_country=df.groupby(['Country'], as_index=False)['Total Revenue'].sum().sort_values(by='Total Revenue',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=revenue_country,x='Country',y= 'Total Revenue')
```

```python
revenue_country=df.groupby(['Country'], as_index=False)['Total Profit'].sum().sort_values(by='Total Profit',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=revenue_country,x='Country',y= 'Total Profit')
```

```python
revenue_country=df.groupby(['Item Type'], as_index=False)['Total Revenue'].sum().sort_values(by='Total Revenue',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=revenue_country,x='Item Type',y= 'Total Revenue')
```

```python
revenue_country=df.groupby(['Sales Channel'], as_index=False)['Total Profit'].sum().sort_values(by='Total Profit',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=revenue_country,x='Sales Channel',y= 'Total Profit')
```

```python
revenue_country=df.groupby(['Region'], as_index=False)['Total Revenue'].sum().sort_values(by='Total Revenue',ascending=False).head(5)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=revenue_country,x='Region',y= 'Total Revenue')
```
